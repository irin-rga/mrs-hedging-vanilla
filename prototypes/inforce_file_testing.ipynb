```python
def create_hdgfctr_lu_df(hdg_dts_df: pd.DataFrame):
    
    hdg_fctr_tbl = hdgfctr_lu_df.merge(hdg_dts_df, on=['HedgeDate'])
    desired_cols = ['HedgeDate', 'CompID', 'HdgFctr']

    hdg_fctr_non_ny_tbl = hdg_fctr_tbl.copy(deep=True)    
    hdg_fctr_non_ny_tbl['CompID'] = 1
    hdg_fctr_non_ny_tbl['HdgFctr'] = hdg_fctr_tbl['HedgeRatio'] * hdg_fctr_tbl['Coins']
    hdg_fctr_non_ny_tbl = hdg_fctr_non_ny_tbl[desired_cols]


    hdg_fctr_ny_tbl = hdg_fctr_tbl.copy(deep=True)    
    hdg_fctr_ny_tbl['CompID'] = 26
    hdg_fctr_ny_tbl['HdgFctr'] = hdg_fctr_tbl['HedgeRatio_NY'] * hdg_fctr_tbl['Coins_NY']
    hdg_fctr_ny_tbl = hdg_fctr_ny_tbl[desired_cols]

    hdgfctr_lu_tbl = pd.concat([hdg_fctr_non_ny_tbl, hdg_fctr_ny_tbl], ignore_index=True)    

    return hdgfctr_lu_tbl
```

```python
def parse_date(date_string, formats):
    """
    Parses a date string using a list of potential formats.

    Args:
        date_string: The string containing the date.
        formats: A list of date format strings.

    Returns:
        A datetime object if parsing is successful, otherwise None.
    """
    for fmt in formats:
        try:
            return datetime.datetime.strptime(date_string, fmt)
        except ValueError:
            continue
    return None

def resolve_inforce_date(inforce_file_path: Optional[Union[str, None]] = None, inforce_file_date: Optional[Union[str, datetime.date, None]] = None):
        """
        Cases:
        1) If inforce_file_date is not None, use it or throw error if it can't be parsed or isn't a date
        2) If inforce_file_date is None, if hedge_file is not None, try and extract date info from file name (check if valid after extracting)
        3) If inforce_file_date and hedge_file are both None, then lookup the hedge date based on the current day and month           
        """            
        inf_dt = None

        if inforce_file_date:
            
            if isinstance(inforce_file_date, datetime.date):
                
                print('inforce_file_date is a date!')
                # Make sure the inforce date is the 1st of the month
                inf_dt = date(inforce_file_date.year, inforce_file_date.month, 1)
                
            elif isinstance(inforce_file_date, str):

                print('inforce_file_date is a str')                
                dt_formats = ["%Y-%m-%d %H:%M:%S", "%m/%d/%Y", "%d %b %Y", "%Y%m%d", "%Y-%m-%d", "%Y-%m-%D"]
                # EG) date_string = "2023-12-11"
                inf_dt = parse_date(inforce_file_date, dt_formats)

                # Make sure the inforce date is the 1st of the month
                inf_dt = date(inf_dt.year, inf_dt.month, 1)
                # inf_date = datetime.strptime(inforce_file_date, date_format).date()
                #   dt.datetime.today().strftime("%Y-%m-%d")
                                        
        else:
             
             if inforce_file_path:
                # Should really check first if this is actually a file and not just a path
                print('inforce_file_date is None but path was entered')
                inf_file_name = os.path.basename(inforce_file_path)
                hdg_mth = int(inf_file_name[:2])
                hdg_yr = int(inf_file_name[3:7])
                inf_dt = date(hdg_yr, hdg_mth, 1)
                return inf_dt
             else:
                print('inforce_file_date is None and path is None')
                # Get the 1st of the month for this month
                inf_dt = datetime.datetime.today()
                inf_dt = date(inf_dt.year, inf_dt.month, 1)

        return inf_dt


test_path = None
test_dt_1 = datetime.datetime.today()
print(resolve_inforce_date(test_path, test_dt_1))

test_path = None
test_dt_1 = date(2024, 12, 17)
print(resolve_inforce_date(test_path, test_dt_1))

test_path = None
test_dt_1 = '2024-12-5'
print(resolve_inforce_date(test_path, test_dt_1))

test_path = None
test_dt_1 = '2024-12-05'
print(resolve_inforce_date(test_path, test_dt_1))

test_path = None
test_dt_1 = '2024125'
print(resolve_inforce_date(test_path, test_dt_1))

test_path = None
test_dt_1 = '20241205'
print(resolve_inforce_date(test_path, test_dt_1))

test_path = None
test_dt_1 = None
print(resolve_inforce_date(test_path, test_dt_1))

test_path = r'C:\Users\S0053071\Repos\Orion_Process_Backup\RGA_Process\12_2024_IUL_Fund_Values_RGA.txt'
test_dt_1 = None
print(resolve_inforce_date(test_path, test_dt_1))
```
